```python
# Importando as bibliotecas

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
```

```python
# Lendo dados tabulares

credit = pd.read_excel("/content/clientes_renamed.xlsx")
```

```python
credit
```

```python
# Agrupando dados

credit_group = credit.groupby(['nivel_escolaridade']).mean().reset_index()
credit_group

# 1=Pós-graduação
# 2=Graduação
# 3=Ensino médio
# 4=Outros
# 5=Desconhecido
# 6=Desconhecido
```

```python
# GRÁFICO COM A MATPLOTLIB

plt.figure(figsize=(8, 6))

plt.xlabel("Nível de escolaridade")
plt.ylabel("Média do valor de crédito concedido")

plt.title("Valor de crédito concedido segundo o nível de escolaridade")

plt.bar(credit_group['nivel_escolaridade'], credit_group['valor_credito'], align='center',
        color='Red', linewidth=1, edgecolor='black');
```

```python
# GRÁFICO COM A SEABORN

fig, axes = plt.subplots(2, 2, figsize=(20,10))

sns.set_style("dark")

sns.barplot(data=credit_group, x="nivel_escolaridade", y="extrato_set_2005", ax=axes[0,0]);
sns.barplot(data=credit_group, x="nivel_escolaridade", y="extrato_ago_2005", ax=axes[0,1]);
sns.barplot(data=credit_group, x="nivel_escolaridade", y="extrato_jul_2005", ax=axes[1,0]);
sns.barplot(data=credit_group, x="nivel_escolaridade", y="extrato_jun_2005", ax=axes[1,1]);
```

```python
# GRÁFICO COM A PLOTLY

fig = go.Figure()

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'], 
    y=credit_group['valor_pag_set_2005'],
    marker_color='red',
    name='Valor pago 09/2005'))

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'],
    y=credit_group['valor_pag_ago_2005'],
    marker_color='yellow',
    name='Valor pago 08/2005'
))

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'],
    y=credit_group['valor_pag_jul_2005'],
    marker_color='green',
    name='Valor pago 07/2005'
))

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'],
    y=credit_group['valor_pag_jun_2005'],
    marker_color='blue',
    name='Valor pago 06/2005'
))

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'],
    y=credit_group['valor_pag_mai_2005'],
    marker_color='crimson',
    name='Valor pago 05/2005'
))

fig.add_trace(go.Bar(
    x=credit_group['nivel_escolaridade'],
    y=credit_group['valor_pag_abr_2005'],
    marker_color='purple',
    name='Valor pago 04/2005'
))

fig.update_layout(
    yaxis=dict(
        title='Valor do pagamento em 2005',
        titlefont_size=16,
        tickfont_size=16,
    ),
    xaxis=dict(
        title='Nível de escolaridade',
        titlefont_size=16,
        tickfont_size=16,
    ),
    legend=dict(
        x=1.0,
        y=1.0,
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ))

fig.show()

# Salvando o gráfico interativo
fig.write_html("valor_pago_escolaridade.html")
```
